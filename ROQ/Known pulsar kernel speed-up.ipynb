Show that using the two basis antenna pattern functions, $a$ and $b$, given in equations 12 and 13 of [Jaranowski, Krolak & Schutz (1998)](http://adsabs.harvard.edu/abs/1998PhRvD..58f3001J) the standard known pulsar search likelihood evaluation can be greatly sped-up.

```python
%matplotlib inline
import numpy as np
from misc import *
import matplotlib.pyplot as plt

from lalapps import pulsarpputils as pppu
```

The full (heterodyned) signal model we have is:
\begin{equation}
h(t) = \frac{h_0}{2}\left[ \left(\frac{1}{2}F_+ (1+\cos{}^2\iota)\cos{\phi_0} + F_{\times}\cos{\iota} \sin{\phi_0 }\right) + i \left(\frac{1}{2}F_+ (1+\cos{}^2\iota)\sin{\phi_0} - F_{\times}\cos{\iota} \cos{\phi_0 }\right)  \right].
\end{equation}
Below we will just define the full model, but the orthonormal basis we use will just require the components that make up the antenna pattern.

```python
def signalmodel(t, h0, cosi, phi0, psi, FA, FB):
  """
  The heterodyned CW signal model.
  """

  ctwopsi = np.cos(2.*psi)
  stwopsi = np.sin(2.*psi)
  fps = FA*ctwopsi + FB*stwopsi
  fcs = FB*ctwopsi - FA*stwopsi
    
  sphi = np.sin(phi0)
  cphi = np.cos(phi0)
  Aplus = 0.5*(1.+cosi**2)
    
  ht = 0.5*h0*((Aplus*fps*cphi + fcs*cosi*sphi) 
        + 1j*(Aplus*fps*sphi - fcs*cosi*cphi))
  return ht
```

We know that the antenna pattern functions (for a given detector and sky position) are defined by
\begin{align}
F_+(t, \psi) &= \sin{\xi}[a(t)\cos{2\psi} + b(t)\sin{2\psi}], \\
F_{\times}(t, \psi) &= \sin{\xi}[b(t)\cos{2\psi} - a(t)\sin{2\psi}],
\end{align}
where $\xi$ is the angle between the detector arms, and is generally $\pi/2$ radians, and from here on we will assume it is subsumed in the $a$ and $b$ terms. Therefore we can write out the signal model to include these full terms. If we just start with the real part we get
\begin{equation}
h_R(t) = \frac{h_0}{2}\left[A_+ \cos{\phi_0} \left(a(t)\cos{2\psi} + b(t)\sin{2\psi}\right) + A_{\times} \sin{\phi_0} \left(b(t)\cos{2\psi} - a(t)\sin{2\psi}\right)\right],
\end{equation}
where $A_+ = (1+\cos{}^2\iota)/2$ and $A_{\times} = \cos{\iota}$.

When calculating a likelihood we need to get the inner products of the data and model, $\sum_{i=1}^N d_R(t_i)h_R(t_i)$ and the model with itself, $\sum_{i=1}^N h_R(t_i)^2$. We can see from above that the summations are only needed over the terms involving $a$ and $b$, and that (for a given sky position and detector) these only need to be computed once. So, for $\sum_{i=1}^N d_R(t_i)h_R(t_i)$, we just require
\begin{equation}
D_{a,R} = \sum_{i=1}^N d_R(t_i)a(t_i) \textrm{ and, } D_{b,R} = \sum_{i=1}^N d_R(t_i)b(t_i),
\end{equation}
which gives
\begin{equation}
\sum_{i=1}^N d_R(t_i)h_R(t_i) = \frac{h_0}{2}\left[A_+ \cos{\phi_0} \left(D_{a,R}\cos{2\psi} + D_{b,R}\sin{2\psi}\right) + A_{\times} \sin{\phi_0} \left(D_{b,R}\cos{2\psi} - D_{a,R}\sin{2\psi}\right)\right],
\end{equation}
where $D_{a,R}$ and $D_{b,R}$ can be precomputed.

Similarly, we need the terms involving the model squared, so
\begin{align}
h_R(t)^2 =& \left(\frac{h_0}{2}\right)^2 \Big\{ A_+^2 \cos{}^2\phi_0 \left[a(t)\cos{2\psi} + b(t)\sin{2\psi} \right]^2 + A_{\times}^2\sin{}^2\phi_0 \left[ b(t)\cos{2\psi} - a(t)\sin{2\psi} \right]^2 \\
 & + 2A_+ A_{\times}\sin{\phi_0}\cos{\phi_0}\left[ a(t)\cos{2\psi} + b(t)\sin{2\psi} \right] \left[ b(t)\cos{2\psi} - a(t)\sin{2\psi} \right] \Big\}.
\end{align}
Therefore this just has sums over
\begin{equation}
A = \sum_{i=1}^N a(t_i)^2, B = \sum_{i=1}^N b(t_i)^2, \textrm{ and } C = 2\sum_{i=1}^N a(t_i)b(t_i),
\end{equation}
giving
\begin{align}
\sum_{i=1}^N h_R(t_i)^2 = & \left(\frac{h_0}{2}\right)^2 \Bigg\{ A_+^2 \cos{}^2\phi_0 \left[ A\cos{}^2 2\psi + B\sin{}^2 2\psi + C\cos{2\psi}\sin{2\psi} \right] \\
& + A_{\times}^2\sin{}^2\phi_0 \left[ B\cos{}^2 2\psi + A\sin{}^2 2\psi - C\cos{2\psi}\sin{2\psi} \right] \\
& + 2A_+ A_{\times}\sin{\phi_0}\cos{\phi_0}\left[ \frac{C}{2}(\cos{}^2 2\psi - \sin{}^2{2\psi}) - A\cos{2\psi}\sin{2\psi} + B\cos{2\psi}\sin{2\psi} \right] \Bigg\}.
\end{align}
Given that the antenna pattern functions in LAL return $F_+$ and $F_{\times}$ we can get $a(t)\sin{\xi} = F_+$ and $b(t)\sin{\xi} = F_{\times}$ back by setting $\psi = 0$.

Similarly, for the imaginary part of the signal we have
\begin{equation}
\sum_{i=1}^N d(t_i)h_I(t_i) = \frac{h_0}{2}\left[A_+ \sin{\phi_0} \left(D_a\cos{2\psi} + D_b\sin{2\psi}\right) - A_{\times} \cos{\phi_0} \left(D_b\cos{2\psi} - D_a\sin{2\psi}\right)\right],
\end{equation}
and
\begin{align}
\sum_{i=1}^N h_I(t_i)^2 = & \left(\frac{h_0}{2}\right)^2 \Bigg\{ A_+^2 \sin{}^2\phi_0 \left[ A\cos{}^2 2\psi + B\sin{}^2 2\psi + C\cos{2\psi}\sin{2\psi} \right] \\
& + A_{\times}^2\cos{}^2\phi_0 \left[ B\cos{}^2 2\psi + A\sin{}^2 2\psi - C\cos{2\psi}\sin{2\psi} \right] \\
& - 2A_+ A_{\times}\sin{\phi_0}\cos{\phi_0}\left[ \frac{C}{2}(\cos{}^2 2\psi - \sin{}^2{2\psi}) - A\cos{2\psi}\sin{2\psi} + B\cos{2\psi}\sin{2\psi} \right] \Bigg\}.
\end{align}

Now let's compare computing the likelihood the slow way with the fast way. The likelihood we will use is the Student's-t likelihood
\begin{align}
p(d|\mathbf{a}) &\propto  \left( \sum_{i=1}^N [d_R(t_i)-h_R(t_i, \mathbf{a})]^2 + [d_I(t_i)-h_I(t_i, \mathbf{a})]^2 \right)^{-N} \\
& = \left( \sum_{i=1}^N d_R(t_i)^2 + h_R(t_i, \mathbf{a})^2 -2d_R(t_i)h_R(t_i, \mathbf{a}) + d_I(t_i)^2 + h_I(t_i, \mathbf{a})^2 -2d_I(t_i)h_I(t_i, \mathbf{a}) \right)^{-N},
\end{align}
which can be calculated by making the appropriate substituions from above.

```python
Ap = 0
Ac = 0
cphi = 0
sphi = 0
c2psi = 0
s2psi = 0
h0_2 = 0

# function for the inner product of the model and the data for speeded up method.
# this assumes that the product of the data and antenna pattern functions a and b
# have aleady been computed
def h_dot_d_fast(h0, cosi, phi0, psi, DaReal, DbReal, DaImag, DbImag):
    global Ap
    global Ac
    global h0_2
    global cphi
    global sphi
    global c2psi
    global s2psi
    
    Ap = 0.5*(1.+cosi**2)
    Ac = cosi
    h0_2 = h0/2.
    cphi = np.cos(phi0)
    sphi = np.sin(phi0)
    c2psi = np.cos(2.*psi)
    s2psi = np.sin(2.*psi)
    
    hdreal = h0_2*(Ap*cphi*( DaReal*c2psi + DbReal*s2psi ) 
                 + Ac*sphi*( DbReal*c2psi - DaReal*s2psi ))
    hdimag = h0_2*(Ap*sphi*( DaImag*c2psi + DbImag*s2psi ) 
                 - Ac*cphi*( DbImag*c2psi - DaImag*s2psi ))

    return hdreal, hdimag

# function for the inner product of the model with itself for the speeded up method.
# This assumes that the products of the antenna pattern functions  a and b have been
# pre-computed
def h_dot_h_fast(A, B, C):
    c2psi2 = c2psi**2
    s2psi2 = s2psi**2
    c2s2 = c2psi*s2psi
    
    hr2 = h0_2**2 * ( (Ap*cphi)**2 * (A*c2psi2 + B*s2psi2 + C*c2s2) +
                      (Ac*sphi)**2 * (B*c2psi2 + A*s2psi2 - C*c2s2) + 
                      2.*Ap*Ac*sphi*cphi*(0.5*C*(c2psi2 -s2psi2) - A*c2s2 + B*c2s2))
    
    hi2 = h0_2**2 * ( (Ap*sphi)**2 * (A*c2psi2 + B*s2psi2 + C*c2s2) +
                      (Ac*cphi)**2 * (B*c2psi2 + A*s2psi2 - C*c2s2) - 
                      2.*Ap*Ac*sphi*cphi*(0.5*C*(c2psi2 -s2psi2) - A*c2s2 + B*c2s2))
    
    return hr2, hi2
    
# function for the inner product of the data - model summing over all the data.
# This requires that the model has just been generated
def hmd_dot_hmd_slow(data, model):
    hdreal = data.real - model.real
    hdimag = data.imag - model.imag
    
    return np.dot(hdreal, hdreal), np.dot(hdimag, hdimag)

# create parameter grid
h0s = np.linspace(0., 5., 5)
cis = np.linspace(-1., 1., 5)
psis = np.linspace(-np.pi/4., np.pi/4., 5)
phi0s = np.linspace(0., 2.*np.pi, 5)

Nevals = len(h0s)*len(cis)*len(psis)*len(phi0s)

import time

# get a set of time scales to test speed-up
timescales = [1., 5., 10., 50., 100., 500.]

fasttimes = []
initimes = []
slowtimes = []

ra = 0. # right ascension at zero radians
dec = 0. # declination at zero radians
det = 'H1' # LIGO Hanford detector

# allocate memory for likelihood grid
lgrid1 = np.ndarray((len(h0s), len(cis), len(psis), len(phi0s)))
lgrid2 = np.ndarray((len(h0s), len(cis), len(psis), len(phi0s)))

for tims in timescales:
    # a time series
    t0 = 0
    tend = 86400.*tims
    N = (tend-t0)/60.
    ts = np.linspace(t0, tend, N)
    dt = ts[1]-ts[0]

    # create data
    data = np.random.randn(len(ts)) + 1j*np.random.randn(len(ts))

    # time the fast method
    t1 = time.time()

    dsumr = np.dot(data.real, data.real)
    dsumi = np.dot(data.imag, data.imag)

    FAs = np.zeros(len(ts))
    FBs = np.zeros(len(ts))
    # create antenna response (without polsaition angle)
    for i, t in enumerate(ts):
        FA, FB = pppu.antenna_response(t, ra, dec, 0.0, det)
        FAs[i] = FA
        FBs[i] = FB
    
    A = np.dot(FAs, FAs)
    B = np.dot(FBs, FBs)
    C = 2.*np.dot(FAs, FBs)

    DaReal = np.dot(FAs, data.real)
    DaImag = np.dot(FAs, data.imag)
    DbReal = np.dot(FBs, data.real)
    DbImag = np.dot(FBs, data.imag)

    eini = time.time()

    for i in range(len(h0s)):
        for j in range(len(cis)):
            for k in range(len(psis)):
                for n in range(len(phi0s)):
                    hdreal, hdimag = h_dot_d_fast(h0s[i], cis[j], phi0s[n], psis[k], DaReal, DbReal, DaImag, DbImag)
                    hr2, hi2 = h_dot_h_fast(A, B, C)
                
                    lgrid1[i, j, k, n] = -N*np.log(dsumr + dsumi + hr2 + hi2 - 2.*hdreal - 2.*hdimag)
                
    e1 = time.time()
    
    fasttimes.append(e1-t1)
    initimes.append(eini-t1)

    # time the slow method
    t2 = time.time()

    for i in range(len(h0s)):
        for j in range(len(cis)):
            for k in range(len(psis)):
                for n in range(len(phi0s)):
                    model = signalmodel(ts, h0s[i], cis[j], phi0s[n], psis[k], FAs, FBs)
                
                    dhr, dhi = hmd_dot_hmd_slow(data, model)
                
                    lgrid2[i, j, k, n] = -N*np.log(dhr + dhi)

    e2 = time.time()
    
    slowtimes.append(e2-t2)

    print "Fast method time %f s" % (e1 - t1)
    print "Fast initialisation time %f s" % (eini - t1)
    print "Fast per-likilihood %f s" % ((e1-eini)/Nevals)
    print "Slow method time %f s" % (e2 - t2)
    print "Slow per-likelihood %f s" % ((e2-t2)/Nevals)
    print "Ratio %f" % ((e2-t2) / (e1-t1))
```

\begin{equation}
p(d|\textbf{a}) = \exp{}\left( -\sum_{i=1}^N \frac{1}{2\sigma_i^2}\left[d_R(t_i)^2 + h_R(t_i, \mathbf{a})^2 -2d_R(t_i)h_R(t_i, \mathbf{a}) + d_I(t_i)^2 + h_I(t_i, \mathbf{a})^2 -2d_I(t_i)h_I(t_i, \mathbf{a}) \right] \right),
\end{equation}
then the varying $\sigma^2$ would need to be incorporated into the calculations of $D_a$, $D_b$, $A$, $B$ and $C$, such that
\begin{equation}
D_a = \sum_{i=1}^N \frac{d(t_i)a(t_i)}{\sigma_i^2} \textrm{ and, } D_b = \sum_{i=1}^N \frac{d(t_i)b(t_i)}{\sigma_i^2},
\end{equation}
and
\begin{equation}
A = \sum_{i=1}^N \left(\frac{a(t_i)}{\sigma_i}\right)^2, B = \sum_{i=1}^N \left(\frac{b(t_i)}{\sigma_i}\right)^2, \textrm{ and } C = 2\sum_{i=1}^N \frac{a(t_i)b(t_i)}{\sigma_i^2}.
\end{equation}

\end{equation}
